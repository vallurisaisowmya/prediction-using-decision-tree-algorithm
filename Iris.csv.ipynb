```python
import numpy as np
import pandas as pd
iris_dataSet= pd.read_csv('Iris.csv')
iris_dataSet 
```

```python
iris_dataSet= iris_dataSet.drop(columns="Id")
iris_dataSet.head()
```

```python
from sklearn import preprocessing
label_encoder=preprocessing.LabelEncoder()
iris_dataSet['Species']=label_encoder.fit_transform(iris_dataSet['Species'])
iris_dataSet
```

```python
#splitting data into dependent and independent variable
x=iris_dataSet.iloc[:,:-1]
x.head()
```

```python
y=iris_dataSet.iloc[:,-1]
y.head()
```

```python
#split the dataset for train and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.33, random_state=42)
```

```python
X_train.head()
```

```python
X_train
```

```python
y_train.shape
```

```python
#DT
from sklearn.tree import DecisionTreeClassifier
tds= DecisionTreeClassifier()
tds.fit(X_train,y_train)
```

```python
#comparision between actual and predicted data
y_predict=tds.predict(X_test)
y_predict
```

```python
y_test=np.array(y_test)
y_test
```

```python
#checking acuracyScore
from sklearn.metrics import accuracy_score
accuracy_score(y_predict,y_test)
```

```python
#confusionMatrix
from sklearn.metrics import confusion_matrix
confusion_matrix(y_predict,y_test)
```

```python
#DT model visualization
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20,17))
dtviz=tree.plot_tree(tds,feature_names=x.columns,filled=True,fontsize=20)
```
